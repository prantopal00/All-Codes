```python

# Load the data
data = data_month

# PART 1: Transformation and Feature Engineering
# Extract substring before the first underscore in PRIMARY_KEY
#data['New_Key'] = data['PRIMARY_KEY'].str.split('_').str[0]

# PART 2: Bucket Calculations and Advanced Feature Engineering

# Replace '.' with '0' in DELSTRING_PAST24M
data['Del_String_New'] = data['DELSTRING_PAST24M'].str.replace('.', '0')

# Split the string into 24 separate columns
del_cols = data['Del_String_New'].str.split(',', expand=True).iloc[:, :24]
del_cols = del_cols.apply(pd.to_numeric, errors='coerce')
del_cols.columns = [f"x{i+1}" for i in range(del_cols.shape[1])]

# Max Bucket Calculations
data['MaxBkt_3M'] = del_cols.iloc[:, :3].max(axis=1)
data['MaxBkt_6M'] = del_cols.iloc[:, :6].max(axis=1)
data['MaxBkt_9M'] = del_cols.iloc[:, :9].max(axis=1)
data['MaxBkt_12M'] = del_cols.iloc[:, :12].max(axis=1)
data['MaxBkt_24M'] = del_cols.max(axis=1)

# Count in Buckets
for bkt in range(1, 4):
    for month in [3, 6, 9, 12, 24,36]:
        data[f'CountInBkt_{bkt}_{month}M'] = (del_cols.iloc[:, :month] == bkt).sum(axis=1)

## Count for 4+ and 1+
for month in [3, 6, 9, 12, 24,36]:
#    data[f'CountInBkt_4Plus_{month}M'] = (del_cols.iloc[:, :month] >= 4).sum(axis=1)
    data[f'CountInBkt_2Plus_{month}M'] = (del_cols.iloc[:, :month] >= 2).sum(axis=1)
    data[f'CountInBkt_3Plus_{month}M'] = (del_cols.iloc[:, :month] >= 3).sum(axis=1)

    data[f'CountInBkt_1Plus_{month}M'] = (del_cols.iloc[:, :month] >= 1).sum(axis=1)

# Ever in Buckets
for bkt in range(1, 4):
    for month in [3, 6, 9, 12, 24]:
        data[f'EverInBkt_{bkt}_{month}M'] = np.where(data[f'CountInBkt_{bkt}_{month}M'] >= 1, 1, 0)

for month in [3, 6, 9, 12, 24]:
    data[f'EverInBkt_3Plus_{month}M'] = np.where(data[f'CountInBkt_3Plus_{month}M'] >= 3, 1, 0)
    data[f'EverInBkt_1Plus_{month}M'] = np.where(data[f'CountInBkt_1Plus_{month}M'] >= 1, 1, 0)

# Latest Delinquency
data['Last_1M_del'] = np.where(del_cols.iloc[:, 0] >= 1, 1, 0)
data['Last_2M_del'] = np.where((del_cols.iloc[:, 0] >= 1) & (del_cols.iloc[:, 1] >= 1), 1, 0)

# Rolling Forward
for period in [1, 2, 3, 6, 12 ,23]:
    roll_fw_cols = [del_cols.iloc[:, i] > del_cols.iloc[:, i+1] for i in range(period)]
    data[f'RollFw{period}'] = sum(roll_fw_cols)

# Rolling Backward
for period in [1, 2, 3, 6, 12, 23]:
    roll_bk_cols = [del_cols.iloc[:, i] < del_cols.iloc[:, i+1] for i in range(period)]
    data[f'Rollbk{period}'] = sum(roll_bk_cols)

# Stable Rolling
for period in [1, 2, 3, 6, 12 ,23]:
    roll_stable_cols = [del_cols.iloc[:, i] == del_cols.iloc[:, i+1] for i in range(period)]
    data[f'Roll_stable_{period}'] = sum(roll_stable_cols)

# Processing DPD
data['DPD_String_New'] = data['DPDSTRING_PAST24M'].str.replace('.', '0')

dpd_cols = data['DPD_String_New'].str.split(',', expand=True).iloc[:, :24]
dpd_cols = dpd_cols.apply(pd.to_numeric, errors='coerce')
dpd_cols.columns = [f"d{i+1}" for i in range(dpd_cols.shape[1])]

# Max DPD Calculations
data['MaxDPD_3M'] = dpd_cols.iloc[:, :3].max(axis=1)
data['MaxDPD_6M'] = dpd_cols.iloc[:, :6].max(axis=1)
data['MaxDPD_9M'] = dpd_cols.iloc[:, :9].max(axis=1)
data['MaxDPD_12M'] = dpd_cols.iloc[:, :12].max(axis=1)
data['MaxDPD_24M'] = dpd_cols.max(axis=1)

# Latest DPD
data['Latest_DPD'] = dpd_cols.iloc[:, 0]
data

# Trend Features
data['TrendBkt_3M'] = del_cols['x1'] - del_cols['x3']
data['TrendBkt_6M'] = del_cols['x1'] - del_cols['x6']
data['TrendDPD_3M'] = dpd_cols['d1'] - dpd_cols['d3']

# Volatility Features
data['Bkt_Std_3M'] = del_cols.iloc[:, :3].std(axis=1)
data['Bkt_Std_6M'] = del_cols.iloc[:, :6].std(axis=1)
data['Bkt_Std_12M'] = del_cols.iloc[:, :12].std(axis=1)
data['DPD_Std_3M'] = dpd_cols.iloc[:, :3].std(axis=1)
data['DPD_Std_6M'] = dpd_cols.iloc[:, :6].std(axis=1)
data['DPD_Std_12M'] = dpd_cols.iloc[:, :12].std(axis=1)

# Recency of High Buckets
def last_occurence(row, bkt_level):
    for i in range(24):
        if row[i] >= bkt_level:
            return i
    return 24

data['LastBkt_1'] = del_cols.apply(lambda row: last_occurence(row, 1), axis=1)
data['LastBkt_2'] = del_cols.apply(lambda row: last_occurence(row, 2), axis=1)
data['LastBkt_3'] = del_cols.apply(lambda row: last_occurence(row, 3), axis=1)
data['LastBkt_4Plus'] = del_cols.apply(lambda row: last_occurence(row, 4), axis=1)

# Max Consecutive Non-Zero / Zero Buckets
def max_consecutive_streak(arr, condition_fn):
    max_streak = streak = 0
    for val in arr:
        if condition_fn(val):
            streak += 1
            max_streak = max(max_streak, streak)
        else:
            streak = 0
    return max_streak

data['MaxConsec_1Plus_Bkts'] = del_cols.apply(lambda row: max_consecutive_streak(row, lambda x: x >= 1), axis=1)
data['MaxConsec_0_Bkts'] = del_cols.apply(lambda row: max_consecutive_streak(row, lambda x: x == 0), axis=1)

# Transitions in Buckets and DPDs
data['Bkt_Transitions_3M'] = del_cols.iloc[:, :3].apply(lambda row: sum(row[i] != row[i+1] for i in range(2)), axis=1)
data['DPD_Transitions_3M'] = dpd_cols.iloc[:, :3].apply(lambda row: sum(row[i] != row[i+1] for i in range(2)), axis=1)

# Lag Features
data['PrevMonth_Bkt'] = del_cols['x2']
data['Prev2Month_Bkt'] = del_cols['x3']

data['AvgBkt_Last3M'] = del_cols.iloc[:, :3].mean(axis=1)
data['AvgBkt_Prev3M'] = del_cols.iloc[:, 3:6].mean(axis=1)
data['Bkt_Change_Recent_vs_Past'] = data['AvgBkt_Last3M'] - data['AvgBkt_Prev3M']
```

```python
from scipy.stats import linregress

data['LastBucketChange'] = del_cols['x1'] - del_cols['x2']
data['EarlyWarningFlag'] = ((del_cols['x1'] > del_cols['x2']) | (del_cols['x2'] > del_cols['x3'])).astype(int)
# Slope over 6 months
data['Del_Slope_6M'] = del_cols.iloc[:, :6].apply(lambda row: linregress(range(6), row)[0], axis=1)

# Slope over 12 months
data['Del_Slope_12M'] = del_cols.iloc[:, :12].apply(lambda row: linregress(range(12), row)[0], axis=1)
weights_6M = np.linspace(1, 2, 6)
weights_12M = np.linspace(1, 2, 12)

data['WeightedAvgBkt_6M'] = del_cols.iloc[:, :6].apply(lambda row: np.average(row, weights=weights_6M), axis=1)
data['WeightedAvgBkt_12M'] = del_cols.iloc[:, :12].apply(lambda row: np.average(row, weights=weights_12M), axis=1)
data['Transitions_12M'] = del_cols.iloc[:, :12].diff(axis=1).fillna(0).ne(0).sum(axis=1)
# Month index (0â€“11) where max bucket occurred in last 12 months
data['MaxBktLag_12M'] = del_cols.iloc[:, :12].apply(lambda row: row.idxmax(), axis=1).str.extract('x(\d+)').astype(float)
```

```python
import numpy as np

def add_bounce_features(df, col="XBktString_L24M"):
    # Convert string â†’ list of ints (store temporarily)
    bounce_lists = df[col].fillna("0,0,"*24).apply(
        lambda x: [int(i) if i.isdigit() else 0 for i in x.split(",")]
    )

    # --- Helper Functions ---
    def max_consec(arr):
        streak = max_streak = 0
        for val in arr:
            if val > 0:
                streak += 1
                max_streak = max(max_streak, streak)
            else:
                streak = 0
        return max_streak

    def last_bounce_recency(arr):
        for i, val in enumerate(arr):
            if val > 0:
                return i
        return len(arr)

    def first_bounce_lag(arr):
        for i in range(len(arr)-1, -1, -1):  # oldest to recent
            if arr[i] > 0:
                return len(arr)-1-i
        return len(arr)

    def avg_spacing(arr):
        idx = [i for i, v in enumerate(arr) if v > 0]
        if len(idx) < 2: return len(arr)
        gaps = [idx[i+1] - idx[i] for i in range(len(idx)-1)]
        return np.mean(gaps)

    def count_transitions(arr):
        return sum(arr[i] != arr[i+1] for i in range(len(arr)-1))

    # --- Core Features (directly added to df) ---
    for m in [3, 6, 12, 24]:
        df[f"BounceCnt_{m}M"] = bounce_lists.apply(lambda x: sum(x[:m]))
        df[f"BounceRate_{m}M"] = df[f"BounceCnt_{m}M"] / m

    df["BounceRatio_RecentVsPast"] = (
        (df["BounceCnt_6M"]+1) / (df["BounceCnt_24M"]-df["BounceCnt_6M"]+1)
    )

    df["EverBounced_12M"] = (df["BounceCnt_12M"] > 0).astype(int)
    df["EverBounced_24M"] = (df["BounceCnt_24M"] > 0).astype(int)

    df["MaxConsecBounce"] = bounce_lists.apply(max_consec)
    df["RecentConsecBounce"] = bounce_lists.apply(lambda x: max_consec(x[:6]))

    df["LastBounceRecency"] = bounce_lists.apply(last_bounce_recency)
    df["FirstBounceLag"] = bounce_lists.apply(first_bounce_lag)
    df["AvgMonthsBetweenBounces"] = bounce_lists.apply(avg_spacing)

    df["BounceVolatility_12M"] = bounce_lists.apply(lambda x: np.std(x[:12]))
    df["BounceTransitions_12M"] = bounce_lists.apply(lambda x: count_transitions(x[:12]))
    df["BounceTrend_12M"] = bounce_lists.apply(lambda x: np.polyfit(range(12), x[:12], 1)[0])

    weights = np.array([0.9**i for i in range(24)])
    df["WeightedBounceScore"] = bounce_lists.apply(lambda x: np.dot(x, weights[:len(x)]))

    df["HighBounceFlag"] = (df["BounceCnt_3M"] >= 2).astype(int)
    df["ChronicBouncer"] = (df["BounceCnt_24M"] >= 12).astype(int)  # >50% months

    print("âœ… Bounce features added directly into data")

# ---- Run on your df ----
add_bounce_features(data, col="XBktString_L24M")

#print(data_month.filter(like="Bounce").head())
```

```python



# Load primary data
#data_month = SCV_DF.copy()
#data_month = pd.read_excel(data_path)
data_month.shape
# Length of DELSTRING_PAST24M_WOCOVID
#length_wo_covid = len(data_month['DELSTRING_PAST24M'][0].split(','))

# Create base DataFrame
base = data_month[['CONNO']]

def get_mode(series):
    return series.value_counts().idxmax() if not series.empty else np.nan

# Prepare DELSTRING data
DelString = data_month['DELSTRING_PAST24M'].str.split(',')

# Initialize metrics
metrics = {
    "MaxBkt_24M": [],
    "MinBkt_24M": [],
    "ModeBkt_24M": [],
    "SDBkt_24M": [],
    "EverInBkt_1_3M": [],
    "EverInBkt_1_6M": [],
    "EverInBkt_1_9M": [],
    "EverInBkt_1_12M": [],
    "EverInBkt_2_3M": [],
    "EverInBkt_2_6M": [],
    "EverInBkt_2_9M": [],
    "EverInBkt_2_12M": [],
    "EverInBkt_3_3M": [],
    "EverInBkt_3_6M": [],
    "EverInBkt_3_9M": [],
    "EverInBkt_3_12M": []
}

# Calculate metrics
for del_list in DelString:
    # Ensure del_list is a list of strings/numbers
    if isinstance(del_list, str):
        del_list = del_list.split(',')  # Split if it's a string
    
    # Convert valid numbers and ignore invalid values (like '.')
    clean_list = [int(x) for x in del_list if x.replace('.', '').isdigit()]

    # Convert to Pandas Series
    del_list = pd.Series(clean_list)  
      # Convert list elements to integers
    metrics["MaxBkt_24M"].append(del_list.max())
    metrics["MinBkt_24M"].append(del_list.min())
    metrics["ModeBkt_24M"].append(get_mode(del_list))
    metrics["SDBkt_24M"].append(del_list.std())
    metrics["EverInBkt_1_3M"].append(1 if (del_list[:3].eq(1)).any() else 0)
    metrics["EverInBkt_1_6M"].append(1 if (del_list[:6].eq(1)).any() else 0)
    metrics["EverInBkt_1_9M"].append(1 if (del_list[:9].eq(1)).any() else 0)
    metrics["EverInBkt_1_12M"].append(1 if (del_list[:12].eq(1)).any() else 0)
    metrics["EverInBkt_2_3M"].append(1 if (del_list[:3].eq(2)).any() else 0)
    metrics["EverInBkt_2_6M"].append(1 if (del_list[:6].eq(2)).any() else 0)
    metrics["EverInBkt_2_9M"].append(1 if (del_list[:9].eq(2)).any() else 0)
    metrics["EverInBkt_2_12M"].append(1 if (del_list[:12].eq(2)).any() else 0)
    metrics["EverInBkt_3_3M"].append(1 if (del_list[:3].eq(3)).any() else 0)
    metrics["EverInBkt_3_6M"].append(1 if (del_list[:6].eq(3)).any() else 0)
    metrics["EverInBkt_3_9M"].append(1 if (del_list[:9].eq(3)).any() else 0)
    metrics["EverInBkt_3_12M"].append(1 if (del_list[:12].eq(3)).any() else 0)

# Add metrics to data_month
metrics_df = pd.DataFrame(metrics)
data_month = pd.concat([data_month, metrics_df], axis=1)


# Final cleanup
data_month.columns = data_month.columns.str.strip()
data_month.rename(columns={'MOB...3': 'MOB'}, inplace=True)

# Print final dataset structure

# Load the data
data = data_month

# PART 1: Transformation and Feature Engineering
# Extract substring before the first underscore in PRIMARY_KEY
#data['New_Key'] = data['PRIMARY_KEY'].str.split('_').str[0]

# PART 2: Bucket Calculations and Advanced Feature Engineering

# Replace '.' with '0' in DELSTRING_PAST24M
data['Del_String_New'] = data['DELSTRING_PAST24M'].str.replace('.', '0')

# Split the string into 24 separate columns
del_cols = data['Del_String_New'].str.split(',', expand=True).iloc[:, :24]
del_cols = del_cols.apply(pd.to_numeric, errors='coerce')
del_cols.columns = [f"x{i+1}" for i in range(del_cols.shape[1])]

# Max Bucket Calculations
data['MaxBkt_3M'] = del_cols.iloc[:, :3].max(axis=1)
data['MaxBkt_6M'] = del_cols.iloc[:, :6].max(axis=1)
data['MaxBkt_9M'] = del_cols.iloc[:, :9].max(axis=1)
data['MaxBkt_12M'] = del_cols.iloc[:, :12].max(axis=1)
data['MaxBkt_24M'] = del_cols.max(axis=1)

# Count in Buckets
for bkt in range(1, 4):
    for month in [3, 6, 9, 12, 24,36]:
        data[f'CountInBkt_{bkt}_{month}M'] = (del_cols.iloc[:, :month] == bkt).sum(axis=1)

## Count for 4+ and 1+
for month in [3, 6, 9, 12, 24,36]:
#    data[f'CountInBkt_4Plus_{month}M'] = (del_cols.iloc[:, :month] >= 4).sum(axis=1)
    data[f'CountInBkt_2Plus_{month}M'] = (del_cols.iloc[:, :month] >= 2).sum(axis=1)
    data[f'CountInBkt_3Plus_{month}M'] = (del_cols.iloc[:, :month] >= 3).sum(axis=1)

    data[f'CountInBkt_1Plus_{month}M'] = (del_cols.iloc[:, :month] >= 1).sum(axis=1)

# Ever in Buckets
for bkt in range(1, 4):
    for month in [3, 6, 9, 12, 24]:
        data[f'EverInBkt_{bkt}_{month}M'] = np.where(data[f'CountInBkt_{bkt}_{month}M'] >= 1, 1, 0)

for month in [3, 6, 9, 12, 24]:
    data[f'EverInBkt_3Plus_{month}M'] = np.where(data[f'CountInBkt_3Plus_{month}M'] >= 3, 1, 0)
    data[f'EverInBkt_1Plus_{month}M'] = np.where(data[f'CountInBkt_1Plus_{month}M'] >= 1, 1, 0)

# Latest Delinquency
data['Last_1M_del'] = np.where(del_cols.iloc[:, 0] >= 1, 1, 0)
data['Last_2M_del'] = np.where((del_cols.iloc[:, 0] >= 1) & (del_cols.iloc[:, 1] >= 1), 1, 0)

# Rolling Forward
for period in [1, 2, 3, 6, 12 ,23]:
    roll_fw_cols = [del_cols.iloc[:, i] > del_cols.iloc[:, i+1] for i in range(period)]
    data[f'RollFw{period}'] = sum(roll_fw_cols)

# Rolling Backward
for period in [1, 2, 3, 6, 12, 23]:
    roll_bk_cols = [del_cols.iloc[:, i] < del_cols.iloc[:, i+1] for i in range(period)]
    data[f'Rollbk{period}'] = sum(roll_bk_cols)

# Stable Rolling
for period in [1, 2, 3, 6, 12 ,23]:
    roll_stable_cols = [del_cols.iloc[:, i] == del_cols.iloc[:, i+1] for i in range(period)]
    data[f'Roll_stable_{period}'] = sum(roll_stable_cols)

# Processing DPD
data['DPD_String_New'] = data['DPDSTRING_PAST24M'].str.replace('.', '0')

dpd_cols = data['DPD_String_New'].str.split(',', expand=True).iloc[:, :24]
dpd_cols = dpd_cols.apply(pd.to_numeric, errors='coerce')
dpd_cols.columns = [f"d{i+1}" for i in range(dpd_cols.shape[1])]

# Max DPD Calculations
data['MaxDPD_3M'] = dpd_cols.iloc[:, :3].max(axis=1)
data['MaxDPD_6M'] = dpd_cols.iloc[:, :6].max(axis=1)
data['MaxDPD_9M'] = dpd_cols.iloc[:, :9].max(axis=1)
data['MaxDPD_12M'] = dpd_cols.iloc[:, :12].max(axis=1)
data['MaxDPD_24M'] = dpd_cols.max(axis=1)

# Latest DPD
data['Latest_DPD'] = dpd_cols.iloc[:, 0]
data

# Trend Features
data['TrendBkt_3M'] = del_cols['x1'] - del_cols['x3']
data['TrendBkt_6M'] = del_cols['x1'] - del_cols['x6']
data['TrendDPD_3M'] = dpd_cols['d1'] - dpd_cols['d3']

# Volatility Features
data['Bkt_Std_3M'] = del_cols.iloc[:, :3].std(axis=1)
data['Bkt_Std_6M'] = del_cols.iloc[:, :6].std(axis=1)
data['Bkt_Std_12M'] = del_cols.iloc[:, :12].std(axis=1)
data['DPD_Std_3M'] = dpd_cols.iloc[:, :3].std(axis=1)
data['DPD_Std_6M'] = dpd_cols.iloc[:, :6].std(axis=1)
data['DPD_Std_12M'] = dpd_cols.iloc[:, :12].std(axis=1)

# Recency of High Buckets
def last_occurence(row, bkt_level):
    for i in range(24):
        if row[i] >= bkt_level:
            return i
    return 24

data['LastBkt_1'] = del_cols.apply(lambda row: last_occurence(row, 1), axis=1)
data['LastBkt_2'] = del_cols.apply(lambda row: last_occurence(row, 2), axis=1)
data['LastBkt_3'] = del_cols.apply(lambda row: last_occurence(row, 3), axis=1)
data['LastBkt_4Plus'] = del_cols.apply(lambda row: last_occurence(row, 4), axis=1)

# Max Consecutive Non-Zero / Zero Buckets
def max_consecutive_streak(arr, condition_fn):
    max_streak = streak = 0
    for val in arr:
        if condition_fn(val):
            streak += 1
            max_streak = max(max_streak, streak)
        else:
            streak = 0
    return max_streak

data['MaxConsec_1Plus_Bkts'] = del_cols.apply(lambda row: max_consecutive_streak(row, lambda x: x >= 1), axis=1)
data['MaxConsec_0_Bkts'] = del_cols.apply(lambda row: max_consecutive_streak(row, lambda x: x == 0), axis=1)

# Transitions in Buckets and DPDs
data['Bkt_Transitions_3M'] = del_cols.iloc[:, :3].apply(lambda row: sum(row[i] != row[i+1] for i in range(2)), axis=1)
data['DPD_Transitions_3M'] = dpd_cols.iloc[:, :3].apply(lambda row: sum(row[i] != row[i+1] for i in range(2)), axis=1)

# Lag Features
data['PrevMonth_Bkt'] = del_cols['x2']
data['Prev2Month_Bkt'] = del_cols['x3']

data['AvgBkt_Last3M'] = del_cols.iloc[:, :3].mean(axis=1)
data['AvgBkt_Prev3M'] = del_cols.iloc[:, 3:6].mean(axis=1)
data['Bkt_Change_Recent_vs_Past'] = data['AvgBkt_Last3M'] - data['AvgBkt_Prev3M']
from scipy.stats import linregress

data['LastBucketChange'] = del_cols['x1'] - del_cols['x2']
data['EarlyWarningFlag'] = ((del_cols['x1'] > del_cols['x2']) | (del_cols['x2'] > del_cols['x3'])).astype(int)
# Slope over 6 months
data['Del_Slope_6M'] = del_cols.iloc[:, :6].apply(lambda row: linregress(range(6), row)[0], axis=1)

# Slope over 12 months
data['Del_Slope_12M'] = del_cols.iloc[:, :12].apply(lambda row: linregress(range(12), row)[0], axis=1)
weights_6M = np.linspace(1, 2, 6)
weights_12M = np.linspace(1, 2, 12)

data['WeightedAvgBkt_6M'] = del_cols.iloc[:, :6].apply(lambda row: np.average(row, weights=weights_6M), axis=1)
data['WeightedAvgBkt_12M'] = del_cols.iloc[:, :12].apply(lambda row: np.average(row, weights=weights_12M), axis=1)
data['Transitions_12M'] = del_cols.iloc[:, :12].diff(axis=1).fillna(0).ne(0).sum(axis=1)
# Month index (0â€“11) where max bucket occurred in last 12 months
data['MaxBktLag_12M'] = del_cols.iloc[:, :12].apply(lambda row: row.idxmax(), axis=1).str.extract('x(\d+)').astype(float)
```
